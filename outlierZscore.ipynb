```
import pandas as pd
import numpy as np
dataset = pd.read_csv("/content/sample_data/heights.csv")
print(dataset)
accending = dataset["height"].values
print(accending)
outliars=[]
#here i'm define a function for z_score
def data_finder(data):
    threshold =1
    #threshold value set 1 or 3 .
    mean = np.mean(data)
    std = np.std(data)
#finding the mean and SD of data
    for i in range(len(data)):
        z_score = (data[i]-mean)/std
        if np.abs(z_score)>threshold:
            outliars.append(i)
    return outliars

data_pred= data_finder(accending)

print(data_pred)
```
